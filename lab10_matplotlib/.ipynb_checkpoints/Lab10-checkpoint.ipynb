```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Read and merge data

```python
df_plant1 = pd.read_csv('../Plant_1_Generation_Data.csv')
df_plant2 = pd.read_csv('../Plant_2_Generation_Data.csv')
df = pd.concat([df_plant1, df_plant2])
df = df.dropna()
df['DATE_TIME'] = pd.to_datetime(df['DATE_TIME']) # dayfirst=True
df = df.set_index(['PLANT_ID', 'SOURCE_KEY', 'DATE_TIME'])
```

```python
df = df.sort_index()
```

```python
df.head()
```

# Analyse of choosen generator

Choosen generators:
    
    1 - 1BY6WEcLGh8j5v7
    
    2 - uHbuxQJl8lW7ozc

```python
generator1_totals = df.loc[4135001.0, '1BY6WEcLGh8j5v7']['DAILY_YIELD']

generator1_totals = generator1_totals.resample('D').agg('last')

generator1_totals = generator1_totals.dropna()

generator1_totals = generator1_totals[generator1_totals>0.0]
```

```python
generator2_totals = df.loc[4135001.0, 'uHbuxQJl8lW7ozc']['DAILY_YIELD']

generator2_totals = generator2_totals.resample('D').agg('last')

generator2_totals = generator2_totals.dropna()

generator2_totals = generator2_totals[generator2_totals>0.0]
```

```python
generator1_totals
```

```python
generator2_totals
```

```python
all_generators_totals = df.groupby(level=2).mean()['DAILY_YIELD']

all_generators_totals = all_generators_totals.resample('D').agg('last')

all_generators_totals = all_generators_totals.dropna()

all_generators_totals = all_generators_totals[all_generators_totals>0.0]
```

```python
all_generators_totals
```

```python
generator1_to_mean = generator1_totals/all_generators_totals

generator1_to_mean = generator1_to_mean.dropna()

generator2_to_mean = generator2_totals/all_generators_totals

generator2_to_mean = generator2_to_mean.dropna()
```

```python
bins = [0, 0.75, 0.85, 0.95, 1.05, 1.15, 1.25, generator1_to_mean.max()] # np.inf
generator1_to_mean = generator1_to_mean.value_counts(bins=bins, sort=False)
```

```python
bins = [0, 0.75, 0.85, 0.95, 1.05, 1.15, 1.25, generator2_to_mean.max()]
generator2_to_mean = generator2_to_mean.value_counts(bins=bins, sort=False)
```
